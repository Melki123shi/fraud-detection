## Business Recommendations

Based on SHAP analysis, here are actionable recommendations for Adey Innovations Inc.:

1. **Transactions within 1 hour of signup should receive additional verification.** This is the strongest fraud signal across both datasets. Implement step-up authentication (SMS/email verification) for purchases made within 60 minutes of account creation.

2. **Block or flag transactions from high-risk countries.** The `is_high_risk_country` feature shows nearly 2x fraud rate. Implement geolocation-based rules that trigger manual review or additional verification for these regions.

3. **Device fingerprinting and velocity limits.** High `users_per_device` ratios indicate fraud rings. Limit the number of unique accounts that can transact from a single device/IP within a rolling window, and flag devices shared by 5+ users.

4. **Night-time transaction monitoring.** Fraud rates peak during early morning hours (0–5 AM). Increase model sensitivity or require additional verification for night-time purchases.

5. **Review PCA features V14, V17, V12, V10 closely for credit card transactions.** These are the top SHAP drivers in the credit dataset. Any downstream rules or alert thresholds should be calibrated around these components.
